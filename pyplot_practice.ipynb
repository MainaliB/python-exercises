```python
import matplotlib.pyplot as plt
import random
```

```python
x1 = list(range(11,20))
x2 = list(range(1,10))
x3 = list(range(21, 30))
plt.plot(x1,x2, c= "green", alpha = 0.7, label = "My line graph")
plt.xlabel("Just some random numbers", rotation = 45)
plt.ylabel("Just some random numbers")
plt.title("Learning how to create Line Graph")
plt.legend()
plt.show()
```

```python
x1 = list(range(11,20))
x2 = list(range(1,10))
x3 = list(range(21, 30))
plt.scatter(x1,x2, s = 10, c = "red", alpha= 0.5)
plt.scatter(x1,x3, s = 50, c = "green", alpha = 0.5)
plt.scatter(x2,x3, s = 35, c = "blue", alpha = 0.5)
plt.show()
```

```python
x1 = [random.random() * 5 for x in range(1,10)]
x1
x2 = [random.randint(1,10) for x in range (1,10)]
x2
x3 = [random.random() for x in range(1,10)]
plt.figure(figsize = (9,9))
plt.scatter(x1,x2, c = "red", s = 100, alpha = 0.6, label = "my  first scatter ")
plt.scatter(x1,x3, c = "green", s = 100, alpha = 0.6, label = "my second scatter")
plt.scatter(x2,x3, c = "blue", s = 100, alpha = 0.65, label = "my third scatter")
plt.title("Me learning how to create scatter plot")
plt.legend()
plt.show()
```

```python
x1 = [random.randint(-5, 25) + n for n in range(150)]
x2 = [x + random.randint(-6, 6) for x in x1]
plt.scatter(range(len(x1)), x1, c = "blue", s = 40, alpha = 0.8, label = "list 1")
plt.scatter(range(len(x2)), x2, c = "red", s = 20, alpha = 0.7, label = "list 2")
plt.legend()
plt.title(" Just making beautiful scatterplot ")
plt.show()
```

```python
plt.scatter(range(len(x1)), x1, c = "red", s = 5, alpha = 1, label = "list 1")
plt.plot(range(len(x1)), x1, ls = '-.',  c = "green", alpha = 1)
plt.show()
```

```python
x = range(1,10)
y = x
y1 = range(10,19)
plt.figure(figsize = (7,7))
plt.scatter(x,y, c = "red", alpha = 0.8, s = 30)
plt.scatter(y,y1, c = "green", alpha = 0.9, s = 40)
plt.yticks([1,2,3,4,5,6, 7, 8, 9], ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 
                                    'eight', 'nine'], rotation = 45)
plt.xticks(rotation = 60)
plt.title("Practicing ticks and rotations")
plt.annotate('Practicing annotation', (4,4), xytext = (6,1), arrowprops = {'arrowstyle': 'wedge', 
                                                                           'facecolor': 'red'})
plt.show()
```

```python
import math
```

```python
x = range(-4, 5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize=(14, 6)) # (width, height)

plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue', label='$\sin(x)$')
plt.plot(x, cos_x, c='orange', label='$\cos(x)$')

plt.legend(loc='upper right')
plt.title('Approximate sin and cosine curves')
plt.xlabel('$x$')

plt.show()
```

```python
x = range(-4, 5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]
plt.figure(figsize = (15,10))
plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue')
plt.plot(x, cos_x, c='orange')
plt.text(-4,1.0, "By bibek", color = "red", fontsize = 10)
plt.annotate("Just a point", (0,0), xytext = (2,0), arrowprops = { 'arrowstyle': '<-', 'facecolor': 'purple'} )
plt.annotate("Just the tip of yello line", (0,1), xytext = (3,1), arrowprops = { 'arrowstyle':'simple',
    'facecolor': 'green'
})
plt.show()
```
